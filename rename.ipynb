```python
import pandas as pd

cs = pd.read_csv("sources.csv")
cs.head()
```

```python
cs['md_name'] = cs['md_name'].str.replace(r'\.pdf$', '.md', regex=True)

# Check the result
print(cs['md_name'].head())
cs.head()
```

```python

import os

dir_files = set(os.listdir("cacx_markdown"))

# Check existence
cs['exists'] = cs['md_name'].isin(dir_files)

# Summary
total = len(cs)
existing_count = cs['exists'].sum()
missing_count = total - existing_count

print(f"Total files in sources.csv : {total}")
print(f"Files existing in cacx_markdown: {existing_count}")
print(f"Files MISSING                : {missing_count}")

# Show missing files
if missing_count > 0:
    print("\n=== MISSING FILES ===")
    missing = cs[~cs['exists']]['md_name'].tolist()
    for i, file in enumerate(missing, 1):
        print(f"{i:2d}. {file}")
else:
    print("\nAll files exist! ✓")
```

```python
# Get actual files in the directory
dir_files = set(os.listdir("cacx_markdown"))

# Mark which ones exist
cs['exists'] = cs['md_name'].isin(dir_files)

# Drop missing ones
cs_clean = cs[cs['exists']].copy().reset_index(drop=True)

# Summary
print(f"Original rows : {len(cs)}")
print(f"Rows after dropping missing: {len(cs_clean)}")
print(f"✅ Successfully kept {len(cs_clean)} files")

# Optional: Drop the helper column
cs_clean = cs_clean.drop(columns=['exists'])

# Preview
cs_clean.head()
```
